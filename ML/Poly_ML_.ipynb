```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://raw.githubusercontent.com/gauravpks/ml-repo/master/Part%202%20-%20Regression/Polynomial%20Regression/Position_Salaries.csv')
df
```

```
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 5)
poly_x = poly_reg.fit_transform(x)
ft = LinearRegression()
ft.fit(poly_x, y)
```

```
poly_x
```

```
y
```

```
ft.predict(poly_reg.fit_transform([[6.5]]))
```

```
plt.scatter(x, y, color = 'red')
plt.plot(x,ft.predict(poly_x),color = 'Black');
```

```
from sklearn import metrics
print('R2: ', metrics.r2_score(y,ft.predict(poly_x)))
```
